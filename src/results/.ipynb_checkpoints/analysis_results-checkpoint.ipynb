```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("beamsearch_results_with_h8.csv")
data
```

```python
#preprocess 

def bs_extract(data):
    
    data["sigma"] = data.apply(lambda x: int(x["sigma"][6:]), axis=1)
    data["filename"] = data.apply(lambda x: (x["filename"][6:] ), axis=1)
    data["m"] = data.apply(lambda x: int(x["filename"].split("_")[0]), axis=1)
    data["n"] = data.apply(lambda x: int(x["filename"].split("_")[1]), axis=1)
    data["index"] = data.apply(lambda x: int(x["filename"].split("_")[-1].split(".")[0]), axis=1)

    data_results=data[["m",  "n", "sigma", "index", "heuristic", "length" ] ] 
    return data_results

def aggregate_bs_results(data_results_grouped):

    data_results_grouped=data_results.groupby(["m", "n", "sigma",  "heuristic" ] ).agg({"length": "mean"}).reset_index()
    data_results_grouped = data_results_grouped.groupby(["heuristic", "n"]).mean().reset_index()[["heuristic", "n", "length"]].sort_values(["n", "heuristic"])
    return data_results 
```

```python
data_results = bs_extract(data)
data_results
```

## Aggregation:


```python
data_results_grouped=data_results.groupby(["m", "n", "sigma",  "heuristic" ] ).agg({"length": "mean"}).reset_index()
data_results_grouped
```

```python
##Visualisations:
print("Results for beam width of 10:\n")
data_results_grouped.groupby("heuristic").mean().reset_index()[["heuristic", "length"]]
```

```python
data_results_grouped.groupby(["heuristic", "n"]).mean().reset_index()[["heuristic", "n", "length"]].sort_values(["n", "heuristic"])
```

# Result for the new IMSBS:


```python
data_ibs = pd.read_csv("imsbs/beamsearch+imbs_results.csv")
data_ibs

data_ibs["sigma"] = data_ibs.apply(lambda x: int(x["sigma"][6:]), axis=1)
data_ibs["filename"] = data_ibs.apply(lambda x: (x["filename"][6:] ), axis=1)
data_ibs["m"] = data_ibs.apply(lambda x: int(x["filename"].split("_")[0]), axis=1)
data_ibs["n"] = data_ibs.apply(lambda x: int(x["filename"].split("_")[1]), axis=1)
data_ibs["index"] = data_ibs.apply(lambda x: int(x["filename"].split("_")[-1].split(".")[0]), axis=1)
data_ibs["alg"] = data_ibs.apply(lambda x: (x["func"].split("_")[0].split(".")[0]), axis=1)


data_ibs_grouped = data_ibs[["alg", "m", "n", "sigma", "beam_width", "heuristic", "length", "time_s" ]].groupby(by=["alg", "m", "n", "sigma", "beam_width", "heuristic" ]).agg( { "length" : "mean", "time_s": "mean"} ).reset_index()

data_grouped_bs = data_ibs_grouped[data_ibs_grouped["alg"] == "beam"]
data_grouped_bs.drop(columns=["alg"], axis = 1, inplace=True)

data_grouped_ibs = data_ibs_grouped[data_ibs_grouped["alg"] == "imbs"]
data_grouped_ibs.drop(columns=["alg"], axis = 1, inplace=True)

```

```python
## further group by for the results of beam_width = 500:

data_grouped_bs_bw_500 = data_grouped_bs[data_grouped_bs["beam_width"] == 500]
data_grouped_ibs_bw_500 = data_grouped_ibs[data_grouped_ibs["beam_width"] == 500]


data_grouped_bs_bw_500.drop(columns=["beam_width"], inplace=True)
data_grouped_ibs_bw_500.drop(columns=["beam_width"], inplace=True)
```

```python

data_grouped_bs_bw_500 = data_grouped_bs_bw_500.rename(columns={ "length" : "BS", "time_s": "BS_time" })
data_grouped_ibs_bw_500 = data_grouped_ibs_bw_500.rename(columns={ "length" : "IBS", "time_s": "IBS_time" })
```

```python
#concat the results:

bs_ibs_beam_500 = data_grouped_bs_bw_500.merge(data_grouped_ibs_bw_500, on=["m", "n", "sigma", "heuristic"], how = "left")
```
