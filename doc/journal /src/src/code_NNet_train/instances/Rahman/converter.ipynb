```python
import os

def list_files_in_directory(directory_path):
    try:
        # Get a list of all files in the directory
        files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
        return files
    except OSError as e:
        print(f"Error listing files in {directory_path}: {e}")
        return []
```

```python
# Example usage:
directory_path = "/home/marko/Desktop/Papers/RLCS-problem/String-inclusion-exclusion-constraints/instances/Rahman/random_StrEc"
files_in_directory = list_files_in_directory(directory_path)

print(f"Files in {directory_path}:")
for file_name in files_in_directory:
    print(file_name)
```

```python
directory_path_out = "/home/marko/Desktop/Papers/RLCS-problem/String-inclusion-exclusion-constraints/instances/Rahman/random_StrEc-converted"

for file_name in files_in_directory:
    f = open(directory_path + "/" + file_name)
    # write line-by-line
    lines = []
    for f_line in f:
        f_line = f_line.replace('\n', '')
        lines.append(f_line)
    
    lines_conversion = []
    sigma_size, p = 20, 0
    #conversion:
    line1 = lines[0] + "    " + f"{sigma_size}" + "    " + f"{p}" + "    " + lines[1]
    lines_conversion.append(line1)
    for line in lines[2:]:
        if len(line)>0:
            lines_conversion.append( str(len(line)) + "    " + line)
        
    print(lines_conversion)    
    #write into a file:
    file_name_write_in = directory_path_out + "/" + file_name
    with open(file_name_write_in, "w") as f_in:
        for line_converted in lines_conversion:
            f_in.write(line_converted +"\n")
    
    f_in.close()
    

```
